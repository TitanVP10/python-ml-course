# La simulación de Monte Carlo

### Estimación de pi

Generamos 2 números aleatorios entre 0 y 1 (x e y)
Calcular x * x + y * y

    Valor inferior a 1 –> Estamos dentro del círculo
    Valor superior a 1 –> Estamos fuera del círculo
    
Calculamos el número total de veces que está dentro del círculo y lo dividimos sobre el número total de intentos para obtener una aproximación de la probabilidad de caer dentro del círculo

Usamos dicha probabilidad para aproximar el valor de pi

Repetimos el experimento para obtener varias aproximaciones de pi 

```python
pi_avg = 0
pi_list = []
n = 1000

for experimento in range(100):
    value = 0
    x = np.random.uniform(0,1,n).tolist()
    y = np.random.uniform(0,1,n).tolist()
    
    for j in range(n):
        z = np.sqrt(x[j]*x[j] + y[j]*y[j])
        if z <= 1:
            value += 1
    float_value = float(value)
    pi_value = float_value * 4 / n
    pi_list.append(pi_value)
    pi_avg += pi_value

pi = pi_avg / 100
print(pi)
plt.plot(pi_list)
```
